## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

## Load the data

Load data from the csv file: <i> 'Categorical.csv'</i>.

```python
raw_data = pd.read_csv('Categorical.csv')
data = raw_data.copy()
```

Remove the duplicate index column from the dataset. 

### There is no duplicate index column in this dataset

## Map the data

Use the <i>'continent'</i> category for this analysis.

```python
data['continent'] = data['continent'].map({'North America':1, 'Asia':2, 'Africa':3, 'Europe':4, 'South America':5,
       'Oceania':6, 'Antarctica':7, 'Seven seas (open ocean)':8})
```

## Select the features

```python
x = data[['Longitude','Latitude','continent']]
```

## Clustering

Use 4 clusters initially.

```python
kmeans = KMeans(9)
kmeans.fit(x)
```

## Clustering results

```python
identified_clusters = kmeans.fit_predict(x)
```

```python
data['cluster'] = identified_clusters
```

## Plot the data

```python
plt.scatter(data['Longitude'],data['Latitude'],c=data['cluster'],cmap='rainbow')
plt.show()
```
